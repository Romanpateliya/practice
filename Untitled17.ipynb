```python
import datetime
def transaction_viewer(tr_obj):
  s = '║'
  for key in tr_obj.keys():
    s += key.center(10) + '\t║\t'
    
  s = s.rstrip()

  print("╔═══════════════","═══════════════════════","═══════════════════════",sep="╦",end="╗")
  print()

  print(s)
  sep = '════════════════╬' #len(s) * 2
  # print(len(sep))

  # for i in range(1,5):
  print("╠═══════════════","═══════════════════════","═══════════════════════",sep="╬",end="╣")

  print()

  print("║", end="")
  for k,v in tr_obj.items():
    if k == 'datetime':
      print(f"{v.strftime('%x %X')}", end='║'.ljust(14))
    else:
      print(f"{v}  ".center(9), end='║'.center(14))
  else:
    print("\n╠═══════════════","═══════════════════════","═══════════════════════",sep="╬",end="╣")
  print()
```
