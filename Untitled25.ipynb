```python
import tkinter
window = tkinter.Tk()
window.geometry("400x200")
window.title("first page")
window.mainloop()

window = tkinter.Tk()
window.geometry("400x200")
lbl = tkinter.label(window , text="fist name")
lbl.place(x=40,y=67)
lbl.mainloop()

window = tkinter.Tk()
window.geometry("400x200+400+200")
ft_lbl = tkinter.lable(window , text="fist name")
ft_lbl.place(x=50,y=70)

f_button = tkinter.entry(window)
f_button.place(x=80,y=90) 

window = tkinter.Tk()
window.geometry()
window.resizeble(False,False)
l_lbl = tkinter.lable(window , text="last name")
l_lbl.place(x=60,y=89)
l_lbl.mainloop()

l_button = tkinter.entry(window)
l_button.place(x=70,y=80)
l_button.mainloop()

button = tkinter.button(window , text="submit")
button.place(x=60,y=85)
button.mainloop()

```
