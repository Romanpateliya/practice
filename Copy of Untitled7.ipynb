```python
 #Write python program that swap two number with temp variable a
numbers = '5,6,7,8,9,9,5'
temp = ''
for n in numbers.split(','):
  if n not in temp:
    temp += n
  else:
    temp = ','.join(temp)
    print(temp)
  
```

```python
'''Write a Python program that will return true ifthe two given integervalues
are equal or their sum or difference is 5'''
def common(list1,list2): 
    result = False
    for x in list1:
      for y in list2:
          if x == y:
            result = True 
            return result
print(common([4,5,7,8],[5,6,7,8]))
print(common([1,2,3,4],[5,6,7,8]))
```
